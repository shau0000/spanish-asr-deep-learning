```python
%matplotlib inline

import torch
import torchaudio
import torch.nn as nn
#import os
import pandas as pd
import librosa
import librosa.display as ds
#import glob
import matplotlib.pyplot as plt
import IPython.display as ipd
import random
import numpy as np



#Audios samples used to create the graphs.

audios = ['nine','six','happy','right','yes','no']

   
    
```

```python
j = 0
plt.figure(figsize=(20,40))
labels = []
features = np.zeros((6,20,44))
k=0
print(features.shape)
for name in audios:
    i = random.randint(1,101)
    filepath = 'Data_Samples/speech_commands_v0.01/' + name + "/" + name +"-" + str(i) + ".wav"
    waveform, sample_rate = librosa.load(filepath)
    print(librosa.get_duration(waveform))
    mfccs = librosa.feature.mfcc(y=waveform, sr=sample_rate)
    if (mfccs.shape[1] != 44):
        print("before " + str(mfccs.shape))
        mfccs = np.resize(mfccs, (20, 44))
    features = np.insert(features, k, mfccs, axis=0)
    k+=1
    labels.append(name)
    print(mfccs.shape)
    j += 1
    plt.subplot(6,2,j)
    plt.title('Waveform for audio = ' + name.capitalize())
    plt.xlabel("Time")
    plt.plot(waveform)
    
    j += 1
    plt.subplot(6,2,j)
    plt.title('MFCCs for audio = ' + name.capitalize())
    plt.xlabel("Time")
    ds.specshow(mfccs, x_axis='time')
    plt.colorbar()
```
